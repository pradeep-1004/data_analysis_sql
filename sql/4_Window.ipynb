```sql
-- Each customer numbering restarts.

select customer_id , order_date , amount , row_number() over(
  partition by customer_id order by amount desc
) as rn from learning_pradeep.new_data.orders_delta
```

```sql
--rank dense rank 


select customer_id , order_date , amount , rank() over(
  partition by amount order by order_date desc 
) as rank
, dense_rank() over(
  partition by amount order by order_date desc 
) as dense_rank 

from learning_pradeep.new_data.orders_delta
```

```sql
--rank dense rank 


select customer_id , order_date , amount , dense_rank() over(
  partition by amount order by order_date desc 
) as dense_rank 

from learning_pradeep.new_data.orders_delta
```

```sql
SELECT customer_id,
       order_id,
       amount,
       SUM(amount) OVER(
            PARTITION BY customer_id
            ORDER BY order_date
            ROWS BETWEEN UNBOUNDED PRECEDING AND CURRENT ROW
       ) AS running_total
FROM learning_pradeep.new_data.orders_delta;
```

```sql
WITH ranked AS (
SELECT *,
       ROW_NUMBER() OVER(
            PARTITION BY customer_id
            ORDER BY order_date DESC
       ) AS row_n
FROM learning_pradeep.new_data.orders_delta
)
SELECT *
FROM ranked
WHERE row_n = 1;
```

```sql
SELECT customer_id,
       order_id,
       amount,
       AVG(amount) OVER(
            PARTITION BY customer_id
            ORDER BY order_date
            ROWS BETWEEN 2 PRECEDING AND CURRENT ROW
       ) AS moving_avg
FROM learning_pradeep.new_data.orders_delta
```
