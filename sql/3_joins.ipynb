```sql
--joins 

--inner

select c.customer_id , c.customer_name , o.order_date , o.amount 

from learning_pradeep.new_data.customers_delta c 
inner join learning_pradeep.new_data.orders_delta o 
on c.customer_id = o.customer_id
```

```sql
--joins 

--left

select c.customer_id , c.customer_name , o.order_date , o.amount 

from learning_pradeep.new_data.customers_delta c 
left join learning_pradeep.new_data.orders_delta o 
on c.customer_id = o.customer_id
```

```sql
--joins 

--right

select c.customer_id , c.customer_name , o.order_date , o.amount 

from learning_pradeep.new_data.customers_delta c 
right join learning_pradeep.new_data.orders_delta o 
on c.customer_id = o.customer_id
```

```sql
--joins 

--full outer join

select c.customer_id , c.customer_name , o.order_date , o.amount 

from learning_pradeep.new_data.customers_delta c 
full outer join learning_pradeep.new_data.orders_delta o 
on c.customer_id = o.customer_id
```

```sql
--joins 

--anti

select c.customer_id , c.customer_name 

from learning_pradeep.new_data.customers_delta c 
anti join learning_pradeep.new_data.orders_delta o 
on c.customer_id = o.customer_id
```
