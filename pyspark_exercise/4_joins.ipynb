```python
df_orders = spark.read.table('orders').display()
```

```python
df_cust = spark.read.table('customers').show()
```

```python
# Example: Check if DataFrames are loaded correctly

assert df_orders is not None, "df_orders is not loaded"
assert df_cust is not None, "df_cust is not loaded"

df_join = df_orders.join(
    df_cust,
    df_orders.customer_id == df_cust.customer_id,
    'inner'
)

display(df_join)
```

```python
df_orders = spark.read.table('orders')
df_cust = spark.read.table('customers')

df_join = df_orders.join(
    df_cust,
    df_orders.customer_id == df_cust.customer_id,
    'inner'
)

display(df_join)
```

```python
df_join = df_orders.join(
    df_cust,
    df_orders.customer_id == df_cust.customer_id,
    'left'
)

display(df_join)
```

```python
df_join = df_orders.join(
    df_cust,
    df_orders.customer_id == df_cust.customer_id,
    'right'
)

display(df_join)
```

```python
df_join = df_orders.join(
    df_cust,
    df_orders.customer_id == df_cust.customer_id,
    'left_semi'
)

display(df_join)
```

```python
df_join = df_orders.join(
    df_cust,
    df_orders.customer_id == df_cust.customer_id,
    'left_anti'
)

display(df_join)
```

```python
products_df = spark.read.table('products')

df_orders.alias("o") \
.join(df_cust.alias("c"), "customer_id", "left") \
.join(products_df.alias("p"), "product_id", "left") \
.select(
    "o.order_id",
    "c.customer_name",
    "c.country",
    "p.category",
    "o.amount"
).show()
```
