```python
df1 = spark.read.table('learning_pradeep.spark_exe.customers')

df1.display()
```

```python
df1.printSchema()
```

```python
df1.select('customer_name','country').display()
```

```python
from pyspark.sql.functions import *
from pyspark.sql.types import * 
```

```python
df1.withColumn('country_upper',lower("country")).show()
```

```python
df1.drop("country").display()
```

```python
#remove duplicates

df1.dropDuplicates(["customer_name"]).display()
```

```python
df1.dropna().show()
```

```python
df1.withColumn('clean_name',trim(lower(col("customer_name")))).show()
```

```python
df1.withColumn('customer_type',
               when(col("country") == ('India'),'domestic')
               .when(col("country") == ('USA'),'international')
               .otherwise('others')
               ).show()
```

```python
df1.filter(
    (col("country").isNotNull()) & 
    (col("customer_name").isNotNull())
).display()
```

```python
df1.filter(
    col("customer_name").isNull()
).show()
```
