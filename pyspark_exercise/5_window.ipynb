```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
orders_df    = spark.read.table("learning_pradeep.spark_exe.orders")
customers_df = spark.read.table("learning_pradeep.spark_exe.customers")
products_df  = spark.read.table("learning_pradeep.spark_exe.products").show()
```

```python
from pyspark.sql.window import Window
```

```python
# row number 

win = Window.partitionBy('customer_id').orderBy(col('amount').desc())

orders_df.withColumn('row_number', row_number().over(win)).show()
```

```python
# rank number 

win = Window.partitionBy('customer_id').orderBy(col('amount').desc())

orders_df.withColumn('rank', rank().over(win)).show()
```

```python
win = Window.partitionBy('customer_id').orderBy(col('amount').desc())

orders_df.withColumn('d_rank', dense_rank().over(win)).show()
```

```python
win = Window.partitionBy('customer_id').orderBy(col('order_date').desc())

df = orders_df.withColumn('d_rank', dense_rank().over(win)).withColumn('rank', rank().over(win)).withColumn('row_number', row_number().over(win))

display(df)
```

```python
win = Window.partitionBy("customer_id").orderBy("order_date") \
           .rowsBetween(Window.unboundedPreceding, Window.currentRow)

orders_df.withColumn("running_total", sum("amount").over(win)).show()
```

```python
win = Window.partitionBy("customer_id").orderBy("order_date")

orders_df \
 .withColumn("prev_order_date", lag("order_date",1).over(win)) \
 .withColumn("gap_days", datediff(col("order_date"), col("prev_order_date"))) \
 .show()
```
