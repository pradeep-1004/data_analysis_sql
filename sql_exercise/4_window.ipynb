```sql
select customer_id , order_date , amount ,

row_number() over(partition by customer_id order by amount desc ) as row_number

from learning_pradeep.sql_exe.fact_orders
```

```sql
select customer_id , order_date , amount ,

row_number() over(partition by customer_id order by amount desc ) as row_number,
rank() over(partition by customer_id order by amount desc ) as rank,
dense_rank() over(partition by customer_id order by amount desc ) as dense_rank

from learning_pradeep.sql_exe.fact_orders
```

```sql
--running total

select customer_id , order_date , amount ,
sum(amount) over(partition by customer_id order by amount 
rows between unbounded preceding and current row
 ) as cum_sum

from learning_pradeep.sql_exe.fact_orders
```

```sql
--Revenue of last 3 orders per customer

select customer_id , order_date , amount ,
sum(amount) over(partition by customer_id order by amount 
rows between 2 preceding and current row
 ) as last_3_orders

from learning_pradeep.sql_exe.fact_orders
```

```sql
-- Get latest order per customer


with ranked as (select *,
rank() over ( partition by customer_id order by order_date desc ) as rank_rn
 from learning_pradeep.sql_exe.fact_orders 
)
select * from ranked where rank_rn = 1
```

```sql
--Remove duplicate customer records, keep latest

with cte as (
  select *,
    rank() over (partition by customer_id order by order_date desc) as dedup
  from learning_pradeep.sql_exe.fact_orders
)
select * from cte where dedup = 1
```

```sql
-- Top 3 highest spending customers per city

with cte as (
  select *,
    rank() over (partition by customer_id order by order_date desc) as top_3
  from learning_pradeep.sql_exe.fact_orders
)
select * from cte where top_3 <= 3
```
