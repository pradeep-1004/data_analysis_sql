```sql
select * from learning_pradeep.sql_exe.fact_orders
```

```sql
select customer_id,
sum(amount) from learning_pradeep.sql_exe.fact_orders 
group by customer_id 
```

```sql
--Revenue by country


select c.country,sum(o.amount) 

from learning_pradeep.sql_exe.fact_orders o
join learning_pradeep.sql_exe.dim_customer c
on o.customer_id = c.customer_id

group by c.country
```

```sql
--Revenue by customer + category


select c.customer_name ,p.product_name , sum(amount)

from learning_pradeep.sql_exe.fact_orders o 
join learning_pradeep.sql_exe.dim_customer c on o.customer_id = c.customer_id

join learning_pradeep.sql_exe.dim_product p on o.product_id = p.product_id

group by c.customer_name,p.product_name
order by c.customer_name
```
