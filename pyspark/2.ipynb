```python
#lower case and upper case

df1.select('customer_id',lower('customer_name').alias('customer_lower'),upper('country').alias('country_upper')).show()
```

```python
df1.select(trim("customer_name").alias("state_clean")).show()
```

```python
df1.show()
```

```python
#contains or like 

df1.filter(col("country").contains('India')).show()
```

```python
df1.filter(col("email").like('%amit%')).show()
```

```python
#sub string 

df1.select(substring('country',1,3).alias('country_code')).show()
```
