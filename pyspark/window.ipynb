```python
from pyspark.sql.types import *
from pyspark.sql.functions import *
from pyspark.sql.window import Window
```

```python
df1 = spark.read.table('learning_pradeep.new_data.customers_delta')

df1.show()
```

```python
df2 = spark.read.table('learning_pradeep.new_data.orders_delta')

df2 = df2.dropna()

df2.show()
```

```python
df2.withColumn('customer_id_r', row_number().over(Window.orderBy('order_date'))).show()
```

```python
df2.withColumn('customer_id_r', rank().over(Window.orderBy('amount'))).show()
```

```python
df2.withColumn('customer_id_r', dense_rank().over(Window.orderBy('amount'))).show()
```

```python
df2.withColumn('cumsum',sum('amount').over(Window.orderBy('product_id'))).show()
```
