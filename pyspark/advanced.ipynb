```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
df1 = spark.read.table('learning_pradeep.new_data.customers_delta')

df1.show()
```

```python
df2 = spark.read.table('learning_pradeep.new_data.orders_delta')

df2.show()
```

```python
joined_df = df2.join(df1, df2.customer_id == df1.customer_id, "inner")
```

```python
joined_df.show()
```

```python
joined_df.select('order_id','customer_name','country','amount').show()
```

```python
left_df = df2.join(df1, df2.customer_id == df1.customer_id, "left")
left_df.select("order_id","customers_delta.customer_id","customer_name","country","amount").show()
```

```python
left_df.show()
```

```python
anti_df = df1.join(df2, df1.customer_id == df2.customer_id, "left_anti")
anti_df.show()
```

#  aggregration 

```python
df2.show()
```

```python
df2.groupBy("customer_id").agg(sum("amount")).show()
```

```python
df2.dropna(subset=["customer_id"]).show()
```

```python
df1.select("country").distinct().show()
```

```python
rev_country = df2.join(df1, df2.customer_id == df1.customer_id, "inner")\
                .groupBy('country')\
                .agg(sum("amount").alias('total_revenue'))

rev_country.show()
```

```python
df2.groupBy("customer_id").agg(count("*").alias("order_count")).show()
```

```python
df2.groupBy("customer_id") \
   .agg(
        count("*").alias("orders"),
        sum("amount").alias("total_spent"),
        avg("amount").alias("avg_spent")
   ).show()
```
